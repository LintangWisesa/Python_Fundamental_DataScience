### Seaborn dataset

- Seaborn Documentation: [https://seaborn.pydata.org/](https://seaborn.pydata.org/)

- Seaborn Dataset: [https://github.com/mwaskom/seaborn-data](https://github.com/mwaskom/seaborn-data)

```python
df = sb.load_dataset('flights')
df.head()
```

```python
# line / scatter plot 
sb.relplot(
    x = 'year', y = 'passengers', data = df, 
    # kind = 'line',
    hue = 'month', size = 'month', 
    # style = 'year'
)
plt.yticks(rotation=90)
plt.ylabel('penumpang')
plt.xlabel('tahun')
plt.grid(True)
plt.title('Flights Data')

plt.show()
```

```python
# bar plot
sb.barplot(
    x = 'year', y = 'passengers', data = df, hue = 'month'
)
plt.legend(loc=0)
plt.show()
```

```python
# cat plot
sb.catplot(
    x = 'year', y = 'passengers', data = df, hue = 'month'
)
plt.show()
```

```python
df.corr() # data numerik
```

```python
sb.heatmap(df.corr(), cmap='hot', annot=True)
plt.show()
```

```python
# heatmap
sb.heatmap(
    df.pivot('month', 'year', 'passengers'),
    annot=True, cmap='hot_r',
    fmt = 'd' # d=int, f=float, .1f=float1dafter.
)
plt.show()
```

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
# violin plot
sb.violinplot(
    x = 'year', y = 'passengers', data = df
)
plt.show()
```
