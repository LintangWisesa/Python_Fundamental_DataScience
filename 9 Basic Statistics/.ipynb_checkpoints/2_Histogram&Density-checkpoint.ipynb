# 7. Histogram

<hr>

- __Histogram__: Show frequencies over intervals, to make it:
    
    1. Decide number of bars & bars width / ranges = $\displaystyle \frac {high - low} {n_ {bars}}$ then round it up.
    2. Decide start value = __(0.5 * unit terkecil)__ before lowest value.
    3. Make a frequency table with ranges.
    4. Build the histogram.

#### _Example_

> Jarak rumah ke sekolah 20 siswa adalah sebagai berikut. Buatlah histogram dengan 5 bars yang merepresentasikan data tersebut.
    __4,6,6,7,11,13,18,18,18,21,24,26,27,35,36,36,42,43,45,49__

#### _Solution_

1. Bar width (range) = $\displaystyle \frac {high - low} {n_ {bars}} = \frac {49 - 4} {5} = 9$, round up to __10__.
2. Unit terkecil = 1, so start value = $\displaystyle low - 0.5 = 4 - 0.5 = $ __3.5__
3. Make a frequency table with ranges & plot the graph!

bins|frequency
---|---
3.5 - 13.5|6
13.5 - 23.5|4
23.5 - 33.5|3
33.5 - 43.5|5
43.5 - 53.5|2

```python
miles = [4,6,6,7,11,13,18,18,18,21,24,26,27,35,36,36,42,43,45,49]
bins = [3.5, 13.5, 23.5, 33.5, 43.5, 53.5]

plt.hist(miles, bins, histtype='stepfilled', rwidth=0.8)
plt.xticks(bins)
plt.xlabel('Miles')
plt.ylabel('Frequency')
plt.title('Home to school distance')
plt.show()
```

<hr>

- __Histogram Shape__
    - __Uniform__: bars about the same
    - __Normal__: taller in the middle, shorter on edges (bell shape)
    - __V-shape__: shorter in middle, taller on edges
    - __Symmetrical__: same on both sides
    - __Skewed right__: extra "stuff" on right
    - __Skewed left__: extra "stuff" on left

```python
data1 = [5,5,5,5,15,15,15,15,25,25,25,25,35,35,35,35,45,45,45,45]
data2 = [5,15,15,15,15,15,25,25,25,25,25,25,25,25,35,35,35,35,45]
data3 = [5,5,5,5,5,15,15,15,15,25,25,35,35,35,35,45,45,45,45,45]
data4 = [5,5,5,15,15,15,15,25,25,25,25,25,25,35,35,35,35,45,45,45]
data5 = [5,5,5,5,15,15,15,15,15,25,25,25,25,35,45,45]
data6 = [5,5,15,25,25,25,25,35,35,35,35,35,45,45,45,45]
bins = [3.5, 13.5, 23.5, 33.5, 43.5, 53.5]

plt.figure(figsize=(15,3))

plt.subplot(161)
plt.hist(data1, bins, histtype='stepfilled', rwidth=0.8)
plt.xticks(bins)
plt.title('Uniform')

plt.subplot(162)
plt.hist(data2, bins, histtype='stepfilled', rwidth=0.8)
plt.xticks(bins)
plt.title('Normal')

plt.subplot(163)
plt.hist(data3, bins, histtype='stepfilled', rwidth=0.8)
plt.xticks(bins)
plt.title('V-Shape')

plt.subplot(164)
plt.hist(data4, bins, histtype='stepfilled', rwidth=0.8)
plt.xticks(bins)
plt.title('Symmetrical')

plt.subplot(165)
plt.hist(data5, bins, histtype='stepfilled', rwidth=0.8)
plt.xticks(bins)
plt.title('Skewed Left')

plt.subplot(166)
plt.hist(data6, bins, histtype='stepfilled', rwidth=0.8)
plt.xticks(bins)
plt.title('Skewed Right')

plt.show()
```

### __Density Curves__: 

Curve that shows overall shape of a data distribution. It very usefull if working with very large population/data sample or very large bin range. 
    
- __Properties of density curve__:
    
    - A density curve must be lie on or above the horizontal axis.
    - Total area under the curve is always equal to 1 or 100%.

```python
a = plt.hist(x, bins, density=True)
plt.plot(a[1][1:] - .05, a[0], 'r-', lw=3)
plt.show()
```

### __Normal Distribution__: 

A special type of density curve that is bell shaped, also known as __bell curve__ or __normal curve__. It describes that most of the data values are near its __mean__. _Specification_: the normal standard distribution is unimodal, has a __single peak__ & symmetric about its mean.

- __Mean__ characterizes the __position__ of the normal distribution curve. Increase the mean, the curve will be moved to right side. Decrease the mean, the curve will be moved to left side.

- __Standard Deviation__ characterizes the __spread__ of the normal distribution curve. The larger standard deviation, the more spread of distribution will be. Standard deviasi juga dapat dijadikan sebagai __interval bins__.

- __Notation__ for population that follow normal standard distribution can be returned like:

    $$X \sim N(\mu, \sigma)$$
    
    for variable $X$, it follows normal distribution $N$ with mean $\mu$ and standard deviation $\sigma$. For example we measure __height__ of all students in a university, then found that it was normally distributed with mean $\mu$ = __55 ft__ & standard deviation $\sigma$ = __0.5 ft__. So we can draw its normal curve with its mean as the middle point & standard deviation as the interval between bins, like this:

```python
lbins = (x.max() + 5.5 - x.min() - 5.5) / 100
min = x.min() + 5.5 - (0.25 * x.min())
max = x.max() + 5.5 + (0.25 * x.min())
bins = np.arange(min, max, lbins)
c = plt.hist(np.random.randn(1000000) + 5.5, bins, density=True)
plt.plot(c[1][1:] - .1, c[0], 'r-', lw=2)
plt.title('Normal Distribution / Bell Curve')
plt.xticks([4, 4.5, 5, 5.5, 6, 6.5, 7])

plt.show()
```

### __The 68-95-99.7 Rule__: 

For example we measure __height__ of all students in a university, then found that it was normally distributed with mean $\mu$ = __55 ft__ & standard deviation $\sigma$ = __0.5 ft__. Mean as the middle point & standard deviation as the interval between bins. The 68-95-99.7 rule is shown as below:

```python
plt.figure(figsize=(12, 6))

lbins = (x.max() + 5.5 - x.min() - 5.5) / 100
min = x.min() + 5.5 - (0.25 * x.min())
max = x.max() + 5.5 + (0.25 * x.min())
bins = np.arange(min, max, lbins)
c = plt.hist(np.random.randn(1000000) + 5.5, bins, density=True, zorder=0)
plt.plot(c[1][1:] - .1, c[0], 'r-', lw=2, zorder=1)
plt.title('The 68-95-99.7 rule')
plt.xticks([4, 4.5, 5, 5.5, 6, 6.5, 7])

# 68%
plt.fill_between([5,6], 0, 0.4, facecolor='y', alpha=0.6, zorder=2)
plt.text(5.3, 0.2, '68%', fontdict={'color':'w', 'family':'impact', 'fontsize':20})
plt.annotate('', xy=(5, .19), xytext=(6, .19), color='w',
             arrowprops=dict(arrowstyle = '<|-|>', facecolor = 'k'))

# 95%
plt.fill_between([4.5,6.5], 0, 0.35, facecolor='pink', alpha=0.6, zorder=1)
plt.text(5.3, 0.12, '95%', fontdict={'color':'w', 'family':'impact', 'fontsize':20})
plt.annotate('', xy=(4.5, .11), xytext=(6.5, .11), color='w',
             arrowprops=dict(arrowstyle = '<|-|>', facecolor = 'k'))

# 99.7%
plt.fill_between([4,7], 0, 0.15, facecolor='lightblue', alpha=0.6, zorder=1)
plt.text(5.3, 0.04, '99.7%', fontdict={'color':'w', 'family':'impact', 'fontsize':20})
plt.annotate('', xy=(4, .03), xytext=(7, .03), color='w',
             arrowprops=dict(arrowstyle = '<|-|>', facecolor = 'k'))

plt.show()
```

- Area antara __-1 standard deviasi__ hingga __+1 standard deviasi__ dari __mean__ (5.0 - 6.0), memiliki nilai __68%__. Artinya 68% populasi/sample berada di range 5.0 - 6.0.

- Area antara __-2 standard deviasi__ hingga __+2 standard deviasi__ dari __mean__ (4.5 - 6.5), memiliki nilai __95%__. Artinya 95% populasi/sample berada di range 4.5 - 6.5.

- Area antara __-3 standard deviasi__ hingga __+3 standard deviasi__ dari __mean__ (4.0 - 7.0), memiliki nilai __99.7%__. Artinya 99.7% populasi/sample berada di range 4.0 - 7.0.
