```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Heatmap

### __1. Visualisasi value or frequency__

```python
df = pd.DataFrame([
    {'sales':'Andi', 'HP A':100, 'HP B':120, 'HP C':65},
    {'sales':'Budi', 'HP A':10, 'HP B':20, 'HP C':5},
    {'sales':'Caca', 'HP A':80, 'HP B':70, 'HP C':60},
])
df = df.set_index('sales')
df
```

```python
plt.imshow(df, cmap='viridis_r')
plt.colorbar()
plt.xticks(np.arange(3), ['HP A', 'HP B', 'HP C'])
# plt.xticks(np.arange(3), df.columns)
plt.yticks(np.arange(3), ['Andi', 'Budi', 'Caca'])
# plt.yticks(np.arange(3), df.index)

for x in range(3):
    for y in range(3):
        plt.text(y-.2, x+.1, df.iloc[x, y], color='w', fontsize=18)

plt.xlim(-.5, 2.5)
plt.ylim(2.5, -.5)
plt.show()
```

### __2. Plot Image or Array as Image__

```python
from PIL import Image
```

```python
# 'L': B/W, 'RGBA': rgba, 'CMYK': cmyk
gbr = Image.open('5.jpg').convert('L')
gbr
```

```python
gbr = np.array(gbr)
print(gbr.shape)
print(gbr)
```

```python
plt.figure(figsize=(8,6))
plt.subplot(121)
plt.imshow(gbr, cmap='hot')
plt.subplot(122)
plt.imshow(gbr, cmap='hot_r')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.subplot(121)
plt.imshow(gbr.T, cmap='Greens')
plt.subplot(122)
plt.imshow(gbr.T[::-1].T, cmap='Greens_r')
plt.show()
```

```python
x = gbr.copy()
# x = x.reshape((100, 400))
plt.imshow(x+x.T[::-1].T, cmap='Greens')
plt.show()
```

### __3. Plot Correlation__

- __Correlation__ tells about the direction & strength of a linear relationship shared between 2 quantitative variables.

```python
# harga daging sapi
df = pd.DataFrame({
    'berat': [10, 20, 30, 40, 50],
    'harga': [1200000, 2000000, 2800000, 3500000, 4000000],
})
df
```

```python
# line plot berat vs harga
# berat = x-axis, Explanatory Variable (explain the outcome of a study), Independent variable (val tidak tergantung variable lain)
# harga = y-axis, Response Variable (measures the outcome of a study), Dependent variable (val tergantung variable lain)

plt.plot(df['berat'], df['harga'], 'ro-')
plt.grid(True)
plt.xlabel('Berat (kg) - Explanatory Var')
plt.ylabel('Harga (Rp) - Response Var')
plt.title('Berat (kg) vs Harga (Rp) Daging Sapi')
plt.show()
```

```python
df.corr(method='pearson')
```

```python
plt.imshow(df.corr())
plt.colorbar()
plt.yticks(np.arange(2), df.columns)
plt.xticks(np.arange(2), df.columns)
for x in range(2):
    for y in range(2):
        plt.text(
            y-.2, x+.1, round(df.corr().iloc[x, y], 3), 
            color='w', fontsize=18
        )
plt.show()
```
