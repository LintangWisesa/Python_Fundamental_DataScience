# 8. Standard Normal Distribution

A special type of normal distribution that has mean $\mu$ = 0 and standard deviation $\sigma$ = 1. Each value on x-axis correspond to __Z-Score__: how far is a data from its mean, measured with standard deviation unit (berapa _standard deviasi_ jauhnya suatu data point dari mean).

```python
x = np.random.randn(1000000)
plt.figure(figsize=(12, 6))

lbins = (x.max() - x.min()) / 100
min = x.min() - (0.5 * x.min())
max = x.max() + (0.5 * x.min())
bins = np.arange(min, max, lbins)
c = plt.hist(x, bins, density=True)
plt.plot(c[1][1:] - .05, c[0], 'r-', lw=2)

# mean
plt.plot((0,0), (0,.4), 'y--')
plt.text(-2.3,0.38,'$\mu$=0 (mid point)')
plt.text(-2.3,0.35,'$\sigma$=1 (interval bin)')

plt.title('Standard Normal Distribution')
plt.show()
```

### __Z-Score Table aka Standard Normal Table__

Selain menggunakan the 68-95-99.7 rule, untuk menentukan % area (probability) di suatu range pada density curve, dapat memanfaatkan Z-Score tabel atau Standard Normal Tabel. Misal: 

- Berapa % area (probability) < 0?
    
    __Answer__: based on Z-Score table, __P(Z<0) = 0.5__

- Berapa % area (probability) P(Z<0.57) dan P(Z>0.57)?

    __Answer__: from Z-Score table, __P(Z<0.57) = 0.7157__ karena __P(Z<0.57) + P(Z>0.57) = 1__ maka didapatkan __P(Z>0.57) = 1 - 0.7157 = 0.2843__.
    
- Berapa nilai Z untuk area dengan probability 0.8461?
    
    __Answer__: based on Z-score table, __P(Z<1.02) = 0.8461__ maka __Z = 1.02__

### __Standardization__

Any normal distribution with various mean $\mu$ & standard deviation $\sigma$ can be transform (or scaling) into standard normal distribution ($\mu$ = 0, $\sigma$ = 1) using __Z-Score formula__:

$$Z = \frac {x - \mu} {\sigma}$$

_**Note**_: 

- Distribusi normal yang belum standard, dapat distandardisasi terlebih dahulu untuk dapat menggunakan tabel Z-Score!

- Z-Score dapat digunakan pula untuk menentuka outlier. High outlier jika value > $\mu + 2 \sigma$. Low outlier jika value < $\mu - 2 \sigma$
