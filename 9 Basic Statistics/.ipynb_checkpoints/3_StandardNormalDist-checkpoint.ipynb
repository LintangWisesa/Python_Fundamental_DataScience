# 8. Standard Normal Distribution

A special type of normal density curve that has mean $\mu$ = 0 and standard deviation $\sigma$ = 1.

```python
lbins = (x.max() - x.min()) / 100
min = x.min() - (0.5 * x.min())
max = x.max() + (0.5 * x.min())
bins = np.arange(min, max, lbins)
c = plt.hist(np.random.randn(1000000), bins, density=True)
plt.plot(c[1][1:] - .05, c[0], 'r-', lw=2)
plt.title('Standard Normal Distribution')
plt.show()
```
