```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
data = {
    'TB': [0.8, 0.85, 0.92, 0.88, 0.82, 0.86, 0.87, 0.93, 0.81, 1],
    'BB': [20, 25, 21, 29, 30, 21, 28, 27, 29, 30]
}
df = pd.DataFrame(data)
df
```

### Standardisasi

Standardisasi $\displaystyle z_{score} = \frac {x - \bar{x}} {s}$ dengan $s$ = standar deviasi (ddof = 0) $\displaystyle s = \sqrt {\frac {\sum (x - \bar{x}) ^ 2} {n-0}} $ .

- High outlier jika $z_{score} > 2.5$

- Low outlier jika $z_{score} < -2.5$

Data terstandarisasi memiliki mean ~ 0 & standar deviasi ~ 1

```python
# std pandas .std() => ddof = 1 sama excel
df['TB'].std()
```

```python
# std numpy np.std() => ddof = 0
np.std(df['TB'])
```

```python
# np.std dengan ddof = 1
np.std(df['TB'], ddof = 1)
```

<hr>

### 1. Standardisasi dengan pandas & numpy

```python
# Standardisasi z score with pandas
# df['Z TB PD'] = (df['TB (m)'] - df['TB (m)'].mean())
# / df['TB (m)'].std()

df['Z TB PD'] = (df['TB'] - df['TB'].mean())/ df['TB'].std()
df['Z BB PD'] = (df['BB'] - df['BB'].mean())/ df['BB'].std()
df
```

```python
# Standardisasi z score with std from numpy
# df['Z TB PD'] = (df['TB (m)'] - df['TB (m)'].mean())
# / np.std(df['TB (m)'])

df['Z TB NP'] = (df['TB'] - df['TB'].mean())/ np.std(df['TB'])
df['Z BB NP'] = (df['BB'] - df['BB'].mean())/ np.std(df['BB'])
df
```

```python
# Standardisasi mean ~ 0 & std ~ 1
print(df['Z TB PD'].mean())
print(df['Z TB NP'].mean())
print(df['Z BB PD'].mean())
print(df['Z BB NP'].mean())

print(df['Z TB PD'].std())
print(df['Z TB NP'].std())
print(df['Z BB PD'].std())
print(df['Z BB NP'].std())
```

Hasil standardisasi dengan numpy & pandas cukup bagus karena Mean ~ 0 & Standar deviasi ~ 1

```python
# standardisasi tidak merubah distribusi data
plt.figure(figsize=(12,4))

plt.subplot(131)
plt.title('TB vs BB')
plt.scatter(df['TB'], df['BB'])

plt.subplot(132)
plt.scatter(df['Z TB PD'], df['Z BB PD'])
plt.title('Z TB pd vs Z BB pd')

plt.subplot(133)
plt.scatter(df['Z TB NP'], df['Z BB NP'])
plt.title('Z TB np vs Z BB np')

plt.show()
```

### 2. Standardisasi dengan sklearn scale

```python
# menggunakan scale
ztb = preprocessing.scale(df['TB'], axis=0)
zbb = preprocessing.scale(df['BB'], axis=0)
df['Z TB sc SKL'] = ztb
df['Z BB sc SKL'] = zbb
df
```

```python
# Standardisasi mean ~ 0 & std ~ 1
print(df['Z TB sc SKL'].mean())
print(df['Z BB sc SKL'].mean())

print(df['Z TB sc SKL'].std())
print(df['Z BB sc SKL'].std())
```

Hasil standardisasi dengan sklearn scale cukup bagus karena Mean ~ 0 & Standar deviasi ~ 1

<hr>

### 3. Standardisasi dengan StandarScaler

- Formula sama dengan standardisasi biasa $\displaystyle z_{score} = \frac {x - \bar{x}} {s}$

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df[['TB', 'BB']])
z = scaler.transform(df[['TB', 'BB']])
z
```

```python
ztb = z[:,0]
zbb = z[:,1]
df['Z TB SKL'] = ztb
df['Z BB SKL'] = zbb
df
```

<hr>

### Inverse Transform

mengembalikan nilai standardisasi ke nilai sebenarnya

```python
# inverse transform
inv = scaler.inverse_transform(
    df[['Z TB SKL', 'Z BB SKL']]
)
print(inv)
```
