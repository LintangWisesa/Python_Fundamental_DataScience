```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import minmax_scale
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import Binarizer
```

```python
data = [169.6,166.8,157.1,181.1,158.4,165.6,166.7,156.5,168.1,165.3]
df = pd.DataFrame(data, columns=['TB'])
df
```

```python
print('Unscaled data', df['TB'])

print('Data after standard scaling', StandardScaler().fit_transform(df[['TB']]))
print('Data after min-max scaling', MinMaxScaler().fit_transform(df[['TB']]))
print('Data after max-abs scaling', MaxAbsScaler().fit_transform(df[['TB']]))
print('Data after robust scaling', RobustScaler(quantile_range=(25, 75)).fit_transform(df[['TB']]))

# print('Data after power transformation (Yeo-Johnson)', PowerTransformer(method='yeo-johnson').fit_transform(df[['TB']]))
# print('Data after power transformation (Box-Cox)', PowerTransformer(method='box-cox').fit_transform(df[['TB']]))

print('Data after quantile transformation (gaussian pdf)', QuantileTransformer(output_distribution='normal').fit_transform(df[['TB']]))
print('Data after quantile transformation (uniform pdf)', QuantileTransformer(output_distribution='uniform').fit_transform(df[['TB']]))

# normalize & normalizer work on row, not col/feature
print('Data after sample-wise L2 normalizing', Normalizer().fit_transform(df[['TB']]))
print('Data after sample-wise L2 normalizing', preprocessing.normalize(df[['TB']]))

print('Data after binarizer:', Binarizer(threshold=167).fit_transform(df[['TB']]))
```

```python
df['Standar'] = StandardScaler().fit_transform(df[['TB']]).reshape(10,)
df['MinMax'] = MinMaxScaler().fit_transform(df[['TB']]).reshape(10,)
df['MaxAbs'] = MaxAbsScaler().fit_transform(df[['TB']]).reshape(10,)
df['Robust'] = RobustScaler(quantile_range=(25, 75)).fit_transform(df[['TB']]).reshape(10,)
# df['Yeo-Johnson'] = PowerTransformer(method='yeo-johnson').fit_transform(df[['TB']]).reshape(10,)
# df['Box-Cox'] = PowerTransformer(method='box-cox').fit_transform(df[['TB']]).reshape(10,)
df['Gaussian pdf'] = QuantileTransformer(output_distribution='normal').fit_transform(df[['TB']]).reshape(10,)
df['Uniform pdf'] = QuantileTransformer(output_distribution='uniform').fit_transform(df[['TB']]).reshape(10,)
df['Normalizer'] = Normalizer().fit_transform(df[['TB']]).reshape(10,)
df['normalize'] = preprocessing.normalize(df[['TB']]).reshape(10,)
df['Binarizer'] = Binarizer(threshold=167).fit_transform(df[['TB']])
df
```
