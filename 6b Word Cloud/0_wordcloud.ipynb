```python
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
```

```python
gbrID = Image.open('indonesia.jpg').convert('RGBA')
gbrID = gbrID.resize((600, 300))
gbrID = np.array(gbrID)
gbrID.shape
```

```python
gbrGE = Image.open('germany.png').convert('RGBA')
gbrGE = gbrGE.resize((600, 300))
gbrGE = np.array(gbrGE)
gbrGE.shape
```

```python
polaID = ImageColorGenerator(gbrID)
polaGE = ImageColorGenerator(gbrGE)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
wc = WordCloud(
    max_font_size=50, max_words=100, background_color="black"
).generate(news)
plt.imshow(wc.recolor(color_func=polaID), interpolation='bilinear')
plt.axis('off')

plt.subplot(122)
wc = WordCloud(
    max_font_size=50, max_words=100, background_color="white"
).generate(news)
plt.imshow(wc.recolor(color_func=polaGE), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
paw = Image.open('sw.png').convert('L')
paw = np.array(paw)
plt.imshow(paw)
plt.show()
```

```python
# 0 => 0 & 255 => 255
def ubah(arr):
    if arr <= 20:
        return 0
    else:
        return 255
    
paw2 = np.ndarray((paw.shape[0], paw.shape[1]), np.int32) 
# np array kosong dg shape == shape paw

for i in range(paw.shape[0]):
    paw2[i] = list(map(ubah, paw[i]))
```

```python
paw2
plt.imshow(paw2)
plt.show()
```

```python
wc = WordCloud(
    max_font_size=100, max_words=100, background_color="black",
    mask = paw2, contour_width=0.2, contour_color='yellow'
).generate(news)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
