```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
# 3D line plot
fig = plt.figure(figsize=(12, 3))
axes = plt.subplot(121, projection='3d')
x = np.arange(1, 11); 
y = np.arange(1, 11); 
z = np.arange(1, 11);
axes.plot_wireframe(x, y, z.reshape(1,-1), color='r', linestyle=':')
plt.show()
```

```python
# 3D scatter plot
fig = plt.figure(figsize=(12, 3))
axes = plt.subplot(121, projection='3d')
x = np.arange(1, 11); 
y = np.arange(1, 11); 
z = np.arange(1, 11);
axes.scatter(x, y, z.reshape(1,-1), color='g', marker='*', s=200)
plt.show()
```

```python
# 3D bar plot
fig = plt.figure(figsize=(12, 3))
axes = plt.subplot(121, projection='3d')
x = np.arange(1, 11); 
y = np.arange(1, 11); 
z = np.full((10,), 8);
lx = np.full((10,), 3); 
ly = np.full((10,), .5); 
tz = np.ones(10)
axes.bar3d(
    x, y, z,    # alas 
    lx, ly, tz,  # lebar xy, tinggi z
    color = ['r', 'g', 'b', 'c', 'm'] * 2
)
axes.set_title(label='3D Bar Plot')
axes.set_xlabel('Nilai X'); axes.set_ylabel('Nilai Y'); axes.set_zlabel('Nilai Z')

plt.show()
```
